```python
# TIC TAC TOE GAME

    
import random


board=[0,1,2,
       3,4,5,
       6,7,8]

def show():
    print (board[0],'|',board[1],'|',board[2])
    print ('---------')
    print (board[3],'|',board[4],'|',board[5])
    print ('---------')
    print (board[6],'|',board[7],'|',board[8])
# Check function

def checkLine(char,spot1,spot2,spot3):
    if board[spot1] == char and board[spot2] == char and board[spot3] == char:
        return True

def checkAll(char):
    #Horizontal lines
    if checkLine(char,0,1,2):
        return True
    if checkLine(char,3,4,5):
        return True
    if checkLine(char,6,7,8):
        return True
    #Vertical lines    
    if checkLine(char,0,3,6):
        return True
    if checkLine(char,1,4,7):
        return True
    if checkLine(char,2,5,8):
        return True
    #Diagonal lines    
    if checkLine(char,0,4,8):
        return True
    if checkLine(char,2,4,6):
        return True
        
        
while True:
    show()
    inp = input("\nSelect as Spot = ")
    inp= int(inp)
    print("\n")
   
    if board[inp]!='x' and board[inp]!='o':
        board[inp]='x'
        
        # Check
        
        if checkAll('x')==True:
            print("\n-~-~X Wins~-~-\n")
            show()
            break;
        
        while True:
            opp = random.randint(0,8)
        
            if  board[opp]!='x' and board[opp]!='o':
                board[opp]='o'
                if checkAll('o')==True:
                    print("\n-~-~O Wins~-~-\n")
                    show()
                    break;
                    
                break;
                
    else:
        print ("Spot is already taken\n")
        
```
