```python
 clean_word_list=[]
for tweet in file['tweets']:
    symbols="!@$%&*()\n_+{}|:\"<>?-=[];',./~`"
    for i in range(0, len(symbols)):
        tweet = tweet.replace(symbols[i],"")
#         tweet = re.sub(r'https.*$', '', tweet)
    if len(tweet)>0:
        #print(word)
        clean_word_list.append(tweet)
file['clean'] = clean_word_list
file.head()
```

```python
hashcount = []
for i in file.tweets:
    j = i.split()
    c = 0 
    for word in j:
        c += len(re.findall('^#', word))
    hashcount.append(c)
file["#_Count"] = hashcount
```
